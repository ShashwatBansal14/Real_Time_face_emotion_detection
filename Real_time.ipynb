```python
import cv2 ## importing opencv
from deepface import DeepFace  ## importing model

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') ## for face detection

cap = cv2.VideoCapture(1)  ## webcam on command
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")
    
while True:
    ret,frame = cap.read()
    enforce_detection =False
    result = DeepFace.analyze(frame,  actions=['emotion'], enforce_detection=False)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  ## cnv img to gray for drawing box
    faces = faceCascade.detectMultiScale(gray,1.1,4)
    
    for(x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
        
    font = cv2.FONT_HERSHEY_SIMPLEX   ## adding font
    
    cv2.putText(frame,
               result[0]['dominant_emotion'],   ## putting things in video
               (50, 50),   ## pixel size
               font, 3,
               (0, 0, 255),
               2,
               cv2.LINE_4)
    cv2.imshow('Demo video',frame)
    
    if cv2.waitKey(4) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
